```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the data
url = "https://raw.githubusercontent.com/Liz283/DataMining/main/heart.csv"
data = pd.read_csv(url)

# Boxplot for 'age'
sns.boxplot(data=data, x='age')
plt.title('Boxplot of Age')
plt.show()  # This displays the current plot

# Boxplot for 'trestbps'
sns.boxplot(data=data, x='trestbps')
plt.title('Boxplot of Resting Blood Pressure (trestbps)')
plt.show()  # This displays the current plot

# Boxplot for 'chol'
sns.boxplot(data=data, x='chol')
plt.title('Boxplot of Cholesterol (chol)')
plt.show()  # This displays the current plot

# Boxplot for 'thalach'
sns.boxplot(data=data, x='thalach')
plt.title('Boxplot of Maximum Heart Rate Achieved (thalach)')
plt.show()  # This displays the current plot

# Boxplot for 'oldpeak'
sns.boxplot(data=data, x='oldpeak')
plt.title('Boxplot of ST Depression Induced by Exercise Relative to Rest (oldpeak)')
plt.show()  # This displays the current plot
```

```python
import pandas as pd

# Load the data
url = "https://raw.githubusercontent.com/Liz283/DataMining/main/heart.csv"
data = pd.read_csv(url)

# Calculate IQR for 'trestbps', 'chol', 'thalach', and 'oldpeak'
variables = ['trestbps', 'chol', 'thalach', 'oldpeak']
outliers_summary = {}

for var in variables:
    Q1_var = data[var].quantile(0.25)
    Q3_var = data[var].quantile(0.75)
    IQR_var = Q3_var - Q1_var

    lower_bound_var = Q1_var - 1.5 * IQR_var
    upper_bound_var = Q3_var + 1.5 * IQR_var

    # Identify outliers
    outliers_var = data[(data[var] < lower_bound_var) | (data[var] > upper_bound_var)]

    # Store summary information
    outliers_summary[var] = {
        "lower_bound": lower_bound_var,
        "upper_bound": upper_bound_var,
        "num_outliers": outliers_var.shape[0]
    }

outliers_summary
```
